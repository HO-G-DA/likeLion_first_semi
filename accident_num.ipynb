```python
import pandas as pd
import numpy as np
```

15세 미만 교통사고 사상자수 및 인구밀도 대비 교통사고율 구하기

```python
#연령별 교통사고 사망자 및 부상자 data import 후 사망자와 부상자 수 합하기

accident_num=pd.read_excel('연령별 교통사고 사상자.xls')
death_num=pd.read_excel('연령별 교통사고 사망자.xls')
accident_total=accident_num['14세이하']+death_num['14세이하']
accident_total
```

```python
#첫줄 총합 및 마지막 NAN 값 정리하고 인덱스 새로 정렬

accident_total2=accident_total[1:26]
accident_total2=accident_total2.reset_index()
accident_total2=accident_total2['14세이하']
accident_total2
```

```python
#구해놨었던 인구수, 인구밀도 data import

seoul_density=pd.read_excel('서울시_인구밀도.xlsx')
seoul_density
```

```python
seoul_density['사고통계']=accident_total2   # 인구밀도 df에 사고통계 컬럼 추가
seoul_density
```

```python
seoul_density['사고비율(인구)']=seoul_density['사고통계']/seoul_density['15세미만인구']  #사고통계에서 인구밀도로 나눠서 인구밀도 대비 사고비율 구하기
seoul_density
```
