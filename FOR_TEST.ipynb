```python
kindergarten=pd.read_csv('num_kindergarten.csv', encoding='euc-kr')
kindergarten
```

```python
new_data={'행정구역별':'세종특별자치시','2019':0}
idx=74
temp1=kindergarten[kindergarten.index <idx]
temp2=kindergarten[kindergarten.index>=idx]
kindergarten = temp1.append(new_data,ignore_index=True).append(temp2, ignore_index=True)
kindergarten
```

```python
kindergarten_num_result=kindergarten['2019']
elementary_num_result=elementary_num['2019']
childcare_num_result=childcare_num['2019']
```

```python
added_num =kindergarten_num_result+elementary_num_result+childcare_num_result
added_num
added_num.rename("added_num")
added_num.name = 'added_num'
```

```python
full_df=pd.concat([new_area,added_num],axis=1)
full_df
```

```python
for i in full_df['added_num']:
    i=int(i)
full_df['added_num']
```

```python
int_area=[]
for i in full_df['2019']:
    i=int(i)
    int_area.append(i)
full_df['int_area']=int_area
full_df['child_density']=full_df['added_num']/full_df['int_area']
full_df['child_density']
```

```python
full_df
```

```python
full_df2=full_df[['소재지(시군구)별(2)','int_area','child_density']]
full_df2
```

```python
full_df2['']
```

```python
q1=full_df2['child_density'].quantile(q=0.25)
q3=full_df2['child_density'].quantile(q=0.75)

qr=q3-q1
qr
```

```python
df_IQR = full_df2[ (full_df2['child_density'] < q3 + qr * 1.5) & (full_df2['child_density'] > q1 - qr * 1.5) ]  

df_IQR
```
